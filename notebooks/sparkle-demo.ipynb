```lean
import Sparkle
open Sparkle.Core.Domain Sparkle.Core.Signal

def counter4 : Signal defaultDomain (BitVec 4) :=
  Signal.circuit do
    let count ← Signal.reg 0#4;
    count <~ count + 1#4;
    return count

#eval! do
  let results := counter4.sample 1
  let nats := results.map BitVec.toNat
  IO.println s!"counter: {nats}"
  Display.waveform "counter4" nats (bitWidth := 4)
```
