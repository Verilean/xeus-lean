# Rich Display in xeus-lean

Lean 4 cells can emit MIME-typed output that JupyterLab renders natively.
Use the sugar commands `#html`, `#latex`, `#md`, `#svg`, `#json`, or call
`Display.html`, `Display.latex`, ... directly from `IO` actions.


## LaTeX math

Where `xeus-lean` shines: Lean 4 is a proof assistant, so equations are
first-class citizens.

````lean
#latex "\\int_0^1 x^2 \\, dx = \\frac{1}{3}"
````

````lean
#latex "\\forall \\varepsilon > 0, \\; \\exists \\delta > 0 : |x - a| < \\delta \\Rightarrow |f(x) - f(a)| < \\varepsilon"
````

## HTML

````lean
#html "<div style='padding: 8px; background: #eef; border-left: 4px solid #44f;'><b>Note:</b> rendered as HTML.</div>"
````

## Markdown

````lean
#md "### Subheading\n\n- bullet **one**\n- bullet *two*\n\n```lean\ndef foo := 42\n```"
````

## SVG

````lean
#svg "<svg xmlns='http://www.w3.org/2000/svg' width='120' height='120'><circle cx='60' cy='60' r='50' fill='#f44' stroke='#222' stroke-width='4'/><text x='60' y='66' text-anchor='middle' font-size='20' fill='white'>Lean</text></svg>"
````

## Programmatic display from `IO`

The same payloads can be emitted from arbitrary `IO` actions, so you can
build them up with regular Lean code.

````lean
#eval do
  for i in [1, 2, 3, 4, 5] do
    Display.latex s!"{i}^2 = {i * i}"
````
