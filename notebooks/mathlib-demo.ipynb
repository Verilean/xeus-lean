# Mathlib in the browser

This notebook loads the on-demand Mathlib bundle.  The default xeus-lean
kernel ships only Init/Std/Lean (~200 MB compressed) so the
first-page boot is small; running the next cell fetches Mathlib + its
deps (Aesop, Batteries, ImportGraph, ProofWidgets, Plausible, Qq) on
top of that.

**Wait for `[%load mathlib] done` to show up in the browser DevTools
console before running the import cell.**  The load runs as fire-and-
forget JS (see CMakeLists.txt for why we can't await it), so the
`%load` cell completes immediately but the actual fetch + decompress
is still in progress in the background.

```lean
%load mathlib
```

Once the load finishes, the next cell should resolve `Mathlib.Tactic`
from the freshly-unpacked oleans.

```lean
import Mathlib.Tactic
```
